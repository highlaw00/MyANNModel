```python
def modify_hour(row):
    # 시간 설정
    date = row['Date Time']
    raw_hour = str(row['Hour Time'])
    hour_time = '0' + raw_hour if len(raw_hour) == 1 else raw_hour
    hour_time = hour_time + ':00'
    date_time = date + ' ' + hour_time
    row['Date Time'] = date_time
    
    return row
```

```python
for i in range(len(seoul_dataframes)):
    seoul_dataframes[i] = seoul_dataframes[i].reset_index().rename(columns={'index': 'Date Time'})
    seoul_dataframes[i] = seoul_dataframes[i].apply(modify_hour, axis=1)
    seoul_dataframes[i] = seoul_dataframes[i].groupby(['Date Time', 'Category']).agg({'Number of Deliveries': list}).reset_index()
seoul_dataframes[0]
```

```python
import datetime

def modify_dataframe(row):
    # 시간 설정
    date_str = row['Date Time'].split(' ')
    hour_str = date_str[1].split(':')
    hour = int(hour_str[0])
    row['Hour Time'] = hour
    
    # 시간의 범주 설정
    hour_time = row['Hour Time']
    if 0 <= hour_time < 6:
        hour_category = 1
    elif 6 <= hour_time < 12:
        hour_category = 2
    elif 12 <= hour_time < 18:
        hour_category = 3
    else:
        hour_category = 4
    row['Hour Category'] = hour_category
    
    # 계절 범주 설정
    date = str(row['Date Time']).split('-')
    month = int(date[1])
    if 3 <= month <= 5:
        season = 1
    elif 6 <= month <= 8:
        season = 2
    elif 9 <= month <= 11:
        season = 3
    else:
        season = 4
    row['Season Category'] = season
    
    # 요일 범주 설정
    raw_str = row['Date Time']
    year = int(raw_str[0:4])
    month = int(raw_str[5:7])
    day = int(raw_str[8:10])
    row['Weekday'] = datetime.date(year, month, day).weekday() + 1
    
    # 공휴일 설정
    date = str(month) + '-' + str(day)
    
    if row['Weekday'] == 6 or row['Weekday'] == 7:
        isHoliday = 1
    else:
        isHoliday = 0
    
    if year == '2019':
        if date in holiday2019:
            isHoliday = 1
    elif year == '2020':
        if date in holiday2020:
            isHoliday = 1
    elif year == '2021':
        if date in holiday2021:
            isHoliday = 1
    row['IsHoliday'] = isHoliday
    
    return row

# seoul_dataframes[0].reset_index(inplace=True)
# seoul_dataframes[0] = seoul_dataframes[0].apply(modify_dataframe, axis=1)
# print(seoul_dataframes[0])

for i, df in enumerate(seoul_dataframes):
    seoul_dataframes[i].reset_index(inplace=True)
    seoul_dataframes[i] = df.apply(modify_dataframe, axis=1)
    print(seoul_dataframes[i])
```

```python
def modify_vector(row):
    vec = row['Vector']
    vec_lis = vec.split(', ')
    cate_lis = ['분식', '족발/보쌈', '찜/탕', '치킨', '카페/디저트', '피자', '한식', '돈까스/일식', '패스트푸드', '아시안/양식']
    dic = {elem: 0 for elem in cate_lis}
    for elem in vec_lis:
        food, num = map(str,elem.split(': '))
        if food in dic:
            dic[food] = int(num)
    row['Vector'] = str(list(dic.values()))[1:-1]
    return row

for i, df in enumerate(seoul_dataframes):
    seoul_dataframes[i] = seoul_dataframes[i].drop(['index'], axis=1)
    seoul_dataframes[i] = seoul_dataframes[i].apply(modify_vector, axis=1)
```

```python
def add_index(row):
    # 체감 온도 추가
    wind_speed = (row['Wind Speed'] * 3.6) ** 0.16
    temp = row['Temperature']
    feeling_idx = 13.12 + 0.6215 * temp - 11.37 * wind_speed + 0.3965 * temp * wind_speed
    row['Feeling Temperature'] = feeling_idx

    # 불쾌 지수 추가
    humidity = row['Humidity']
    discomfort_idx = 0.81 * temp + 0.01 * humidity * (0.99 * temp - 14.3) + 46.3
    if 80 <= discomfort_idx:
        res = 4
    elif 75 <= discomfort_idx < 80:
        res = 3
    elif 68 <= discomfort_idx < 75:
        res = 2
    else:
        res = 1
    row['Discomfort Index'] = res
    
    dust = row['Dust Density']
    if 151 <= dust:
        res = 4
    elif 81 <= dust < 151:
        res = 3
    elif 31 <= dust < 80:
        res = 2
    else:
        res = 1
    row['Dust Level'] = res
    
    code = row['Weather Code']
    res = 0
    if pd.isna(code):
        code = 0
    else:
        code = int(code)
        # 토큰을 만들며 확인합니다.
        while code:
            token = code % 100
            if 1 <= token <= 4:
                res = 1
                break
            code = code // 100
    row['IsRain'] = res
    return row

for i in range(len(seoul_dataframes)):
    seoul_dataframes[i] = seoul_dataframes[i].apply(add_index, axis=1)
```
