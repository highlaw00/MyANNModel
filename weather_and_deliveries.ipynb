```python
import pandas as pd

data = pd.read_csv("C:\\Users\\choiyool00\\Desktop\\Co-op\\ann\\csvs\\deli3.csv", names=['Hour Time', 'Category', 'State', 'City', 'Number of Deliveries' ])
seoul = data.query('State == "서울특별시"')
```

```python
temp_data = pd.read_csv("C:\\Users\\choiyool00\\Desktop\\Co-op\\ann\\csvs\\surface2019_2019.csv", encoding='euc-kr')
temp_data
```

```python
seoul.reset_index(inplace=True)

def modify_datetime(row):
    date = row['index']
    raw_hour = str(row['Hour Time'])
    hour_time = '0' + raw_hour if len(raw_hour) == 1 else raw_hour
    hour_time = hour_time + ':00'
    date_time = date + ' ' + hour_time
    row['index'] = date_time
    return row

seoul = seoul.apply(modify_datetime, axis=1)
```

```python
seoul = seoul.rename(columns={'index': 'Date Time'})
```

```python
temp_data = temp_data[['일시', '기온(°C)', '강수량(mm)', '습도(%)', '적설(cm)', '현상번호(국내식)']]
temp_data = temp_data.rename(columns={'일시': 'Date Time'})
```

```python
merged = pd.merge(seoul, temp_data, on='Date Time')
```
