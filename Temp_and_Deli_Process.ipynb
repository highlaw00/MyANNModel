```python
import pandas as pd
import datetime

dfs = []
for i in range(3):
    title = "C:\\Users\\choiyool00\\Desktop\\Co-op\\ann\\merged_temp_and_deli" + str(2019 + i) + ".csv"
    dfs.append(pd.read_csv(title))
    print(dfs[i])
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].rename(columns = {'기온(°C)': 'Temperature','습도(%)':'Humidity', '현상번호(국내식)':'Weather Code', '강수량(mm)': 'Precipitation'})
    dfs[i] = dfs[i].rename(columns = {'풍속(m/s)': 'Wind Speed', '일조(hr)': 'Sunshine', '전운량(10분위)': 'Cloud Level'})
dfs[0]
```

```python
dfs[0] = dfs[0].groupby(['Date Time', 'Category', 'Temperature', 'Humidity', 'Wind Speed', 'Sunshine', 'Precipitation', 'Cloud Level', 'Weather Code', 'Hour Category', 'Season Category', 'Weekday', 'IsHoliday']).agg({'Number of Deliveries': list}).reset_index()
dfs[0]
```

```python
for i in range(len(dfs)):
    dfs[i]['Vector'] = dfs[i].apply(lambda row: ', '.join([f"{c}: {v}" for c, v in zip(row['Category'].split(', '), row['Total Deliveries'].split(', '))]), axis=1)
    dfs[i] = dfs[i].drop(columns={'Category', 'Total Deliveries'})
```

```python
def add_index(row):
    # 체감 온도 추가
    wind_speed = (row['Wind Speed'] * 3.6) ** 0.16
    temp = row['Temperature']
    feeling_idx = 13.12 + 0.6215 * temp - 11.37 * wind_speed + 0.3965 * temp * wind_speed
    row['Feeling Temperature'] = feeling_idx

    # 불쾌 지수 추가
    humidity = row['Humidity']
    discomfort_idx = 0.81 * temp + 0.01 * humidity * (0.99 * temp - 14.3) + 46.3
    if 80 <= discomfort_idx:
        res = 4
    elif 75 <= discomfort_idx < 80:
        res = 3
    elif 68 <= discomfort_idx < 75:
        res = 2
    else:
        res = 1
    row['Discomfort Index'] = res
    
    dust = row['Dust Density']
    if 151 <= dust:
        res = 4
    elif 81 <= dust < 151:
        res = 3
    elif 31 <= dust < 80:
        res = 2
    else:
        res = 1
    row['Dust Level'] = res
    
    return row

for i in range(len(dfs)):
    dfs[i] = dfs[i].apply(add_index, axis=1)
```

```python
def modify_vector(row):
    vec = row['Vector']
    vec_lis = vec.split(', ')
    cate_lis = ['분식', '족발/보쌈', '찜/탕', '치킨', '카페/디저트', '피자', '한식', '돈까스/일식', '패스트푸드', '아시안/양식']
    dic = {elem: 0 for elem in cate_lis}
    for elem in vec_lis:
        food, num = map(str,elem.split(': '))
        dic[food] = int(num)
    row['Vector'] = str(list(dic.values()))[1:-1]
    return row
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].apply(modify_vector, axis=1)
dfs[0]
```

```python
def modify_else(row):
    row['Weekday'] = int(row['Weekday'])
    code = row['Weather Code']
    res = 0 # res가 0이면 특별 기후 없음, 1이면 비, 2이면 눈, 3이면 흐림입니다.
    if pd.isna(code):
        code = 0
    else:
        code = int(code)
        # 토큰을 만들며 확인합니다.
        while code:
            token = code % 100
            if 1 <= token <= 4:
                res = 1
                break
            code = code // 100
    row['IsRain'] = res
    return row
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].apply(modify_else, axis=1)
dfs[0]
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].reset_index().drop(columns=['Weather Code', 'Date Time'])
dfs[0]
```

```python
for i in range(len(dfs)):
    title = './Vectorized_Seoul_Datas' + str(2019 + i) + '.csv'
    dfs[i].to_csv(title)
```
