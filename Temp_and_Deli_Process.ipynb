```python
import pandas as pd
import datetime

data = pd.read_csv("C:\\Users\\choiyool00\\Desktop\\Co-op\\ann\\merged_csvs\\merged_temp_and_deli2021.csv")
data
```

```python
Weekday = []
for index, row in data.iterrows():
    raw_str = row['Date Time']
    year = int(raw_str[0:4])
    month = int(raw_str[5:7])
    day = int(raw_str[8:10])
    Weekday.append(datetime.date(year, month, day).weekday())
data['Weekday'] = Weekday
```

```python
data = data.rename(columns={'기온(°C)': 'Temperature','습도(%)':'Humidity', '현상번호(국내식)':'Weather Code'})
data
```

```python
data = data.drop(['강수량(mm)', '적설(cm)'], axis = 1)
data
```

```python
result = data.groupby(['Date Time','Hour Time','Temperature', 'Humidity', 'Weather Code', 'Weekday', 'Category'])['Number of Deliveries'].agg(list).reset_index()
result['Total Deliveries'] = result['Number of Deliveries'].apply(sum)
result = result.groupby(['Date Time', 'Hour Time']).agg({'Category': lambda x: ', '.join(x), 'Total Deliveries': lambda x: ', '.join(map(str, x)), 'Temperature': 'mean', 'Humidity': 'mean', 'Weekday': 'mean', 'Weather Code': 'mean'})
```

```python
result['Vector'] = result.apply(lambda row: ', '.join([f"{c}: {v}" for c, v in zip(row['Category'].split(', '), row['Total Deliveries'].split(', '))]), axis=1)
result = result.drop(columns={'Category', 'Total Deliveries'})
```

```python
# 카테고리 3개 이하 주문 자르기
result = result[result['Vector'].apply(lambda x: len(x.split(', ')) >= 2)]
result
```

```python
def modify_vector(row):
    vec = row['Vector']
    vec_lis = vec.split(', ')
    cate_lis = ['분식', '야식', '족발/보쌈', '찜/탕', '치킨', '카페/디저트', '피자', '한식', '회', '돈까스/일식', '패스트푸드', '아시안/양식', '중식']
    dic = {elem: 0 for elem in cate_lis}
    for elem in vec_lis:
        food, num = map(str,elem.split(': '))
        dic[food] = int(num)
    row['Vector'] = str(list(dic.values()))[1:-1]
    return row
```

```python
result = result.apply(modify_vector, axis=1)
```

```python
def modify_else(row):
    row['Weekday'] = int(row['Weekday'])
    code = row['Weather Code']
    res = 0 # res가 0이면 특별 기후 없음, 1이면 비, 2이면 눈, 3이면 흐림입니다.
    if pd.isna(code):
        code = 0
    else:
        code = int(code)
        # 토큰을 만들며 확인합니다.
        while code:
            token = code % 100
            if 1 <= token <= 4:
                res = 1
                break
            elif 5 <= token <= 15 or token == 22:
                res = 2
                break
            elif code == 92:
                res = 3
                break
            code = code // 100
    row['Code'] = res
    return row
```

```python
result = result.apply(modify_else, axis=1)
```

```python
result = result.reset_index().drop(columns=['Weather Code', 'Date Time'])
result
```

```python
result.to_csv('./Vectorized_Seoul_Datas2021.csv')
```
